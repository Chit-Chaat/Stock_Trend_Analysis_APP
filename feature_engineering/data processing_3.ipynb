```python
import yfinance as yf
import os
import pandas as pd
from datetime import datetime

start_date = "2020-01-01"
end_date = datetime.today()
data_df = yf.download( "FB", start=start_date, end=end_date)[['Open', 'Close', 'Low', 'High', 'Volume']]
data_df.reset_index(inplace=True)
data_df = data_df.rename(columns={"Open": "Stock_Open", "Close": "Stock_Close", "Low": "Stock_Low","High": "Stock_High","Volume": "Stock_Volume"})
```

```python
DATA_DIR = "./DATA"
DOW_FILE_PATH = "Dow_daily.csv"
NASDAQ_FILE_PATH = "Nasdaq_daily.csv"
SP500_FILE_PATH = "SP500_daily.csv"

dow_df = pd.read_csv(os.path.join(DATA_DIR, DOW_FILE_PATH))[['Open', 'Close', 'Low', 'High']]
dow_df = dow_df.rename(columns={"Open": "Dow_Open", "Close": "Dow_Close", "Low": "Dow_Low","High": "Dow_High"})
```

```python
concat_df = pd.concat([data_df, dow_df.reindex(data_df.index)], axis=1)
```

```python
concat_df
```

```python
nasdaq_df = pd.read_csv(os.path.join(DATA_DIR, NASDAQ_FILE_PATH))[['Open', 'Close', 'Low', 'High']]
nasdaq_df = nasdaq_df.rename(columns={"Open": "Nasdaq_Open", "Close": "Nasdaq_Close", "Low": "Nasdaq_Low","High": "Nasdaq_High"})
concat_df = pd.concat([concat_df, nasdaq_df.reindex(concat_df.index)], axis=1)
```

```python
sp500_df = pd.read_csv(os.path.join(DATA_DIR, SP500_FILE_PATH))[['Open', 'Close', 'Low', 'High']]
sp500_df = sp500_df.rename(columns={"Open": "SP500_Open", "Close": "SP500_Close", "Low": "SP500_Low","High": "SP500_High"})
concat_df = pd.concat([concat_df, sp500_df.reindex(concat_df.index)], axis=1)
```
